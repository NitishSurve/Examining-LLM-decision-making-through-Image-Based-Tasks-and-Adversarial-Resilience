```python
import glob
import pandas as pd

# change path for different models
# path = 'img_data/gpt/gpt-4-vision/*.csv'   ## ChatGPT
# path = 'img_data/gemini/gemini-1.5-flash/*.csv'   ## Gemini

# path = 'img_data/gpt/img_gpt-4-vision/*.csv'   ## ChatGPT
path = 'img_data/gemini/img_gemini-1.5-flash/*.csv'   ## Gemini

csv_files = glob.glob(path)
```

```python
#combined_df = pd.concat([pd.read_csv(file) for file in csv_files], ignore_index=True)

combined_df = pd.DataFrame()
for file in csv_files:
    df = pd.read_csv(file)
    combined_df = pd.concat([combined_df,df])
combined_df.rename(columns={
    'run' : 'id'
}, inplace=True)
combined_df
```

```python
# output_path = 'img_data/gpt_4_vision.csv'   ## ChatGPT
# output_path = 'img_data/gemini.csv'   ## Gemini

# output_path = 'img_data/img_gpt_4_vision.csv'   ## img ChatGPT
output_path = 'img_data/img_gemini.csv'   ## Gemini
combined_df.to_csv(output_path)
```
